```python
import requests as r
import pandas as pd
import random
```

```python
X =[]
def get_weather_data(dict):
  try:
    country = dict['sys']['country']
    #city =
  except KeyError:
    country= 'Not avail.'
    #city = None
  features=[(dict['main']['temp']-273.15),dict['main']['pressure'],dict['main']['humidity'],dict['wind']['speed'],dict['clouds']['all'],country,dict['weather'][0]['main']]
  features=pd.DataFrame(features)
  features=features.T
  X.append(features)
  return X
```

```python
API_KEY = '0dcff6d1048e25ef4de7711889a9917e'
#lat = random.uniform(-60, 60)
#lon = random.uniform(-180,180)
for i in range(300):
  lat = random.uniform(-60, 60)
  lon = random.uniform(-180,180)
  url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}"
  dict = r.get(url)
  dict=dict.json()
  get_weather_data(dict)
X=pd.concat(X)
X.columns=['Temp','Pressure','Humidity','Wind_speed','Cloudiness', 'Country','Target']
X = X.reset_index(drop=True)
X.sample(6)
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Target'] = le.fit_transform(X['Target'])
```

```python
#Splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X.iloc[:,4], X['Target'],
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
# Reshape X_train to be 2D for a single feature
X_train_reshaped = X_train.values.reshape(-1, 1)
rf.fit(X_train_reshaped,y_train)
```

```python
from sklearn.metrics import accuracy_score
# Reshape X_test to be 2D for a single feature
X_test_reshaped = X_test.values.reshape(-1, 1)
y_pred = rf.predict(X_test_reshaped)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
